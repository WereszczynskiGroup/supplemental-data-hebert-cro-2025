```python
import numpy as np
import matplotlib.pyplot as plt

# Load data into dictionaries
WTM = {}
A3M = {}
A8M = {}

for i in range(1, 6):  # M1–M5
    WTM[str(i)] = np.loadtxt(f'WT_M{i}_c_contacts.dat', unpack=True, comments='#', usecols=(1))
    A3M[str(i)] = np.loadtxt(f'A3_M{i}_c_contacts.dat', unpack=True, comments='#', usecols=(1))
    A8M[str(i)] = np.loadtxt(f'A8_M{i}_c_contacts.dat', unpack=True, comments='#', usecols=(1))

datasets = [WTM, A3M, A8M]
labels = ['Wild Type', 'Act 3', 'Act 8']
x = np.arange(1, len(WTM["1"]) + 1)

fig, axs = plt.subplots(5, 3, figsize=(6, 6), sharex=True, sharey=True)

for col, data in enumerate(datasets):  # 0 = WTM, 1 = A3M, 2 = A8M
    for row in range(5):  # 0–4 for M1–M5
        y = data[str(row + 1)]
        axs[row, col].plot(x/10, y)
        #if col == 0:
            #axs[row, col].set_ylabel(f'Contacts')
        if row == 0:
            axs[row, col].set_title(labels[col])

fig.supxlabel('Time (ns)')
fig.supylabel('Contacts')
#plt.tight_layout()
plt.suptitle('Monomer C-terminus 1-50 Contacts vs. Time')
plt.savefig('monomer_cterm_time.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data into dictionaries
WTdim1, WTdim2 = {}, {}
A3dim1, A3dim2 = {}, {}
A8dim1, A8dim2 = {}, {}

for i in range(1, 6):  # M1–M5
    WTdim1[str(i)] = np.loadtxt(f'WTdim{i}_c_contacts.dat', unpack=True, comments='#', usecols=(1))
    WTdim2[str(i)] = np.loadtxt(f'WTdim{i}_c_contacts.dat', unpack=True, comments='#', usecols=(2))
    
    A3dim1[str(i)] = np.loadtxt(f'A3dim{i}_c_contacts.dat', unpack=True, comments='#', usecols=(1))
    A3dim2[str(i)] = np.loadtxt(f'A3dim{i}_c_contacts.dat', unpack=True, comments='#', usecols=(2))
    
    A8dim1[str(i)] = np.loadtxt(f'A8dim{i}_c_contacts.dat', unpack=True, comments='#', usecols=(1))
    A8dim2[str(i)] = np.loadtxt(f'A8dim{i}_c_contacts.dat', unpack=True, comments='#', usecols=(2))

# Combine for easy access
dim1_sets = [WTdim1, A3dim1, A8dim1]
dim2_sets = [WTdim2, A3dim2, A8dim2]
labels = ['Wild Type', 'Act 3', 'Act 8']

# Assume all have same x length
x = np.arange(1, len(WTdim1["1"]) + 1)

# Create plot
fig, axs = plt.subplots(5, 3, figsize=(6, 6), sharex=True, sharey=True)

for col in range(3):  # WT, A3, A8
    for row in range(5):  # M1–M5
        dim1 = dim1_sets[col][str(row + 1)]
        dim2 = dim2_sets[col][str(row + 1)]
        axs[row, col].plot(x, dim1, label='Subunit 1', color='blue', alpha=0.5)
        axs[row, col].plot(x, dim2, label='Subunit 2', color='orange', alpha=0.5)
        
        if row == 0:
            axs[row, col].set_title(labels[col])

fig.supxlabel('Time (ns)')
fig.supylabel('Contacts')
plt.suptitle('Dimer C-terminus Contacts vs. Time')
# Create a single legend to the right of all subplots
handles = [
    plt.Line2D([0], [0], color='blue', alpha=0.5, label='Subunit 1'),
    plt.Line2D([0], [0], color='orange', alpha=0.5, label='Subunit 2')
]

fig.legend(handles=handles, loc='center left', bbox_to_anchor=(1.01, 0.5))

plt.tight_layout()
plt.savefig('dimer_cterm_time.pdf')
```
