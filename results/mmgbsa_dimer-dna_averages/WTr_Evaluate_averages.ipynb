```python
import re
import sys
import numpy as np
import statistics as stat
from pymbar import timeseries
from math import sqrt

a8_1 = './WTr/WTrgbsa1.xvg'
a8_2 = './WTr/WTrgbsa2.xvg' 
a8_3 = './WTr/WTrgbsa3.xvg'
a8_4 = './WTr/WTrgbsa4.xvg'
a8_5 = './WTr/WTrgbsa5.xvg'

output_file1 = "WTr_complex_data.txt"
output_file2 = "WTr_receptor_data.txt"
output_file3 = "WTr_ligand_data.txt"
```

```python

time1 = np.loadtxt(a8_1, unpack=True,comments=('#'),delimiter=',',usecols=0)
bond1 = np.loadtxt(a8_1, unpack=True,comments=('#'),delimiter=',',usecols=1)
angle1 = np.loadtxt(a8_1, unpack=True,comments=('#'),delimiter=',',usecols=2)
dihed1 = np.loadtxt(a8_1, unpack=True,comments=('#'),delimiter=',',usecols=3)
vdwaals1 = np.loadtxt(a8_1, unpack=True,comments=('#'),delimiter=',',usecols=7)
eel1 = np.loadtxt(a8_1, unpack=True,comments=('#'),delimiter=',',usecols=8)
vdw_1_4_1 = np.loadtxt(a8_1, unpack=True,comments=('#'),delimiter=',',usecols=9)
eel_1_4_1 = np.loadtxt(a8_1, unpack=True,comments=('#'),delimiter=',',usecols=10)
egb1 = np.loadtxt(a8_1, unpack=True,comments=('#'),delimiter=',',usecols=11)
esurf1 = np.loadtxt(a8_1, unpack=True,comments=('#'),delimiter=',',usecols=12)
total1 = np.loadtxt(a8_1, unpack=True,comments=('#'),delimiter=',',usecols=15)

time2 = np.loadtxt(a8_2, unpack=True,comments=('#'),delimiter=',',usecols=0)
bond2 = np.loadtxt(a8_2, unpack=True,comments=('#'),delimiter=',',usecols=1)
angle2 = np.loadtxt(a8_2, unpack=True,comments=('#'),delimiter=',',usecols=2)
dihed2 = np.loadtxt(a8_2, unpack=True,comments=('#'),delimiter=',',usecols=3)
vdwaals2 = np.loadtxt(a8_2, unpack=True,comments=('#'),delimiter=',',usecols=7)
eel2 = np.loadtxt(a8_2, unpack=True,comments=('#'),delimiter=',',usecols=8)
vdw_1_4_2 = np.loadtxt(a8_2, unpack=True,comments=('#'),delimiter=',',usecols=9)
eel_1_4_2 = np.loadtxt(a8_2, unpack=True,comments=('#'),delimiter=',',usecols=10)
egb2 = np.loadtxt(a8_2, unpack=True,comments=('#'),delimiter=',',usecols=11)
esurf2 = np.loadtxt(a8_2, unpack=True,comments=('#'),delimiter=',',usecols=12)
total2 = np.loadtxt(a8_2, unpack=True,comments=('#'),delimiter=',',usecols=15)

time3 = np.loadtxt(a8_3, unpack=True,comments=('#'),delimiter=',',usecols=0)
bond3 = np.loadtxt(a8_3, unpack=True,comments=('#'),delimiter=',',usecols=1)
angle3 = np.loadtxt(a8_3, unpack=True,comments=('#'),delimiter=',',usecols=2)
dihed3 = np.loadtxt(a8_3, unpack=True,comments=('#'),delimiter=',',usecols=3)
vdwaals3 = np.loadtxt(a8_3, unpack=True,comments=('#'),delimiter=',',usecols=7)
eel3 = np.loadtxt(a8_3, unpack=True,comments=('#'),delimiter=',',usecols=8)
vdw_1_4_3 = np.loadtxt(a8_3, unpack=True,comments=('#'),delimiter=',',usecols=9)
eel_1_4_3 = np.loadtxt(a8_3, unpack=True,comments=('#'),delimiter=',',usecols=10)
egb3 = np.loadtxt(a8_3, unpack=True,comments=('#'),delimiter=',',usecols=11)
esurf3 = np.loadtxt(a8_3, unpack=True,comments=('#'),delimiter=',',usecols=12)
total3 = np.loadtxt(a8_3, unpack=True,comments=('#'),delimiter=',',usecols=15)

time4 = np.loadtxt(a8_4, unpack=True,comments=('#'),delimiter=',',usecols=0)
bond4 = np.loadtxt(a8_4, unpack=True,comments=('#'),delimiter=',',usecols=1)
angle4 = np.loadtxt(a8_4, unpack=True,comments=('#'),delimiter=',',usecols=2)
dihed4 = np.loadtxt(a8_4, unpack=True,comments=('#'),delimiter=',',usecols=3)
vdwaals4 = np.loadtxt(a8_4, unpack=True,comments=('#'),delimiter=',',usecols=7)
eel4 = np.loadtxt(a8_4, unpack=True,comments=('#'),delimiter=',',usecols=8)
vdw_1_4_4 = np.loadtxt(a8_4, unpack=True,comments=('#'),delimiter=',',usecols=9)
eel_1_4_4 = np.loadtxt(a8_4, unpack=True,comments=('#'),delimiter=',',usecols=10)
egb4 = np.loadtxt(a8_4, unpack=True,comments=('#'),delimiter=',',usecols=11)
esurf4 = np.loadtxt(a8_4, unpack=True,comments=('#'),delimiter=',',usecols=12)
total4 = np.loadtxt(a8_4, unpack=True,comments=('#'),delimiter=',',usecols=15)

time5 = np.loadtxt(a8_5, unpack=True,comments=('#'),delimiter=',',usecols=0)
bond5 = np.loadtxt(a8_5, unpack=True,comments=('#'),delimiter=',',usecols=1)
angle5 = np.loadtxt(a8_5, unpack=True,comments=('#'),delimiter=',',usecols=2)
dihed5 = np.loadtxt(a8_5, unpack=True,comments=('#'),delimiter=',',usecols=3)
vdwaals5 = np.loadtxt(a8_5, unpack=True,comments=('#'),delimiter=',',usecols=7)
eel5 = np.loadtxt(a8_5, unpack=True,comments=('#'),delimiter=',',usecols=8)
vdw_1_4_5 = np.loadtxt(a8_5, unpack=True,comments=('#'),delimiter=',',usecols=9)
eel_1_4_5 = np.loadtxt(a8_5, unpack=True,comments=('#'),delimiter=',',usecols=10)
egb5 = np.loadtxt(a8_5, unpack=True,comments=('#'),delimiter=',',usecols=11)
esurf5 = np.loadtxt(a8_5, unpack=True,comments=('#'),delimiter=',',usecols=12)
total5 = np.loadtxt(a8_5, unpack=True,comments=('#'),delimiter=',',usecols=15)

###########################################################################################################################
###### These values assume that your ligand, receptor, and complex all had the same number of frames calculated on! #######
######### If the number of frames is different for ligand, complex, or receptor, your calculations will be off. ###########
comp0 = 0
compf = int(len(time1)/3)
rec0 = compf 
recf = int(2*len(time1)/3)
lig0 = recf
ligf = int(len(time1))
###########################################################################################################################
###########################################################################################################################

internal1 = bond1 + angle1 + dihed1
Elec1 = eel1 + egb1 + eel_1_4_1
VdW1 = vdwaals1 + vdw_1_4_1 + esurf1

internal2 = bond2 + angle2 + dihed2
Elec2 = eel2 + egb2 + eel_1_4_2
VdW2 = vdwaals2 + vdw_1_4_2 + esurf2

internal3 = bond3 + angle3 + dihed3
Elec3 = eel3 + egb3 + eel_1_4_3
VdW3 = vdwaals3 + vdw_1_4_3 + esurf3

internal5 = bond5 + angle5 + dihed5
Elec5 = eel5 + egb5 + eel_1_4_5
VdW5 = vdwaals5 + vdw_1_4_5 + esurf5

comp04 = 0
compf4 = int(len(time4)/3)
rec04 = compf4 
recf4 = int(2*len(time4)/3)
lig04 = recf4
ligf4 = int(len(time4))

internal4 = bond4 + angle4 + dihed4
Elec4 = eel4 + egb4 + eel_1_4_4
VdW4 = vdwaals4 + vdw_1_4_4 + esurf4



######
## Concatenate everything into longer time series ### 
VdWcom = np.concatenate((VdW1[comp0:compf], VdW2[comp0:compf], VdW3[comp0:compf], VdW4[comp04:compf4], VdW5[comp0:compf]))
VdWrec = np.concatenate((VdW1[rec0:recf], VdW2[rec0:recf], VdW3[rec0:recf], VdW4[rec04:recf4], VdW5[rec0:recf]))
VdWlig = np.concatenate((VdW1[lig0:ligf], VdW2[lig0:ligf], VdW3[lig0:ligf], VdW4[lig04:ligf4], VdW5[lig0:ligf]))

Eleccom = np.concatenate((Elec1[comp0:compf], Elec2[comp0:compf], Elec3[comp0:compf], Elec4[comp04:compf4], Elec5[comp0:compf]))
Elecrec = np.concatenate((Elec1[rec0:recf], Elec2[rec0:recf], Elec3[rec0:recf], Elec4[rec04:recf4], Elec5[rec0:recf]))
Eleclig = np.concatenate((Elec1[lig0:ligf], Elec2[lig0:ligf], Elec3[lig0:ligf], Elec4[lig04:ligf4], Elec5[lig0:ligf]))

internalcom = np.concatenate((internal1[comp0:compf], internal2[comp0:compf], internal3[comp0:compf], internal4[comp04:compf4], internal5[comp0:compf]))
internalrec = np.concatenate((internal1[rec0:recf], internal2[rec0:recf], internal3[rec0:recf], internal4[rec04:recf4], internal5[rec0:recf]))
internallig = np.concatenate((internal1[lig0:ligf], internal2[lig0:ligf], internal3[lig0:ligf], internal4[lig04:ligf4], internal5[lig0:ligf]))
######

totalcom = Eleccom + VdWcom + internalcom

with open(output_file1, "w") as f:
    f.write("# Complex\n # Internal Elec VdW\n")  # Header
    for i in range(len(VdWcom)):
        f.write(f"{i} {internalcom[i]:.6f} {Eleccom[i]:.6f} {VdWcom[i]:.6f}\n")
print(f"Data written to {output_file1}")

with open(output_file2, "w") as f:
    f.write("# Receptor\n # Internal Elec VdW\n")  # Header
    for i in range(len(VdWrec)):
        f.write(f"{i} {internalrec[i]:.6f} {Elecrec[i]:.6f} {VdWrec[i]:.6f}\n")
print(f"Data written to {output_file2}")

with open(output_file3, "w") as f:
    f.write("# Ligand\n # Internal Elec VdW\n")  # Header
    for i in range(len(VdWlig)):
        f.write(f"{i} {internallig[i]:.6f} {Eleclig[i]:.6f} {VdWlig[i]:.6f}\n")
print(f"Data written to {output_file3}")

internalcom, Eleccom, VdWcom = np.loadtxt(f'{output_file1}', usecols=(1,2,3), unpack=True, comments=('#'))
internalrec, Elecrec, VdWrec = np.loadtxt(f'{output_file2}', usecols=(1,2,3), unpack=True, comments=('#'))
internallig, Eleclig, VdWlig = np.loadtxt(f'A8_ligand_data.txt', usecols=(1,2,3), unpack=True, comments=('#'))

totalcom = Eleccom + VdWcom + internalcom

# Define sum of squares function for propagating error
def sumsq(a,b,c):
    sumsquares = a*a + b*b + c*c
    return sumsquares

SI = timeseries.statistical_inefficiency(totalcom)
N = np.floor(len(totalcom)/SI)
print('This inefficiency test is based on the time series of the complex total energy.')
print('The number of significant points is '+str(N)+'. Decorrelation time = '+str(len(totalcom)/N)+' frames \n')

avgVdW = np.average(VdWcom) - np.average(VdWrec) - np.average(VdWlig)

avgEl = np.average(Eleccom) - np.average(Elecrec) - np.average(Eleclig)

avginternal = np.average(internalcom) - np.average(internalrec) - np.average(internallig)

avgtotal = avgEl + avginternal + avgVdW

VdWsterr = sqrt(sumsq(np.std(VdWcom)/sqrt(N),np.std(VdWlig)/sqrt(N),np.std(VdWrec)/sqrt(N))) 
r_avgVdW = np.round(avgVdW, decimals=1)
r_VdWsterr = np.round(VdWsterr, decimals=1)


Elsterr = sqrt(sumsq(np.std(Eleccom)/sqrt(N),np.std(Eleclig)/sqrt(N),np.std(Elecrec)/sqrt(N))) 
r_avgEl = np.round(avgEl, decimals=1)
r_Elsterr = np.round(Elsterr, decimals=1)


internalsterr = sqrt(sumsq(np.std(internalcom)/sqrt(N),np.std(internallig)/sqrt(N),np.std(internalrec)/sqrt(N))) 
r_avginternal = np.round(avginternal, decimals=1)
r_internalsterr = np.round(internalsterr, decimals=1)


totalsterr = sqrt(sumsq(VdWsterr,Elsterr,internalsterr))
r_avgtotal = np.round(avgtotal,decimals=1)
r_totalsterr = np.round(totalsterr,decimals=1)

print('WTr_VdW: '+str(r_avgVdW)+' +/- '+str(r_VdWsterr)+'\n')
print('WTr_Elec: '+str(r_avgEl)+' +/- '+str(r_Elsterr)+'\n')
print('WTr_internal: '+str(r_avginternal)+' +/- '+str(r_internalsterr)+'\n')
print('WTr_Total: '+str(r_avgtotal)+' +/- '+str(r_totalsterr)+'\n')
```
