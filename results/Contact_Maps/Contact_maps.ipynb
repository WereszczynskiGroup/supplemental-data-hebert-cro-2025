```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import Normalize

# Load data
wtdimmat = np.loadtxt("WTdim_contmap.dat")[:, 1:]
wtruncdimmat = np.loadtxt("WTdna_contmap.dat")[:, 1:]

A8dimmat = np.loadtxt("A8dim_contmap.dat")[:, 1:]
T8dimmat = np.loadtxt("A8dna_contmap.dat")[:, 1:]

A3dimmat = np.loadtxt("A3dim_contmap.dat")[:, 1:]
T3dimmat = np.loadtxt("A3dna_contmap.dat")[:, 1:]

# Compute difference maps
diff_A3 = wtdimmat - A3dimmat
diff_A8 = wtdimmat - A8dimmat
diff_T3 = wtruncdimmat - T3dimmat
diff_T8 = wtruncdimmat - T8dimmat

# Get residue index ranges
top_residue_indices = np.arange(1, wtdimmat.shape[0] + 1)
bottom_residue_indices = np.arange(1, wtruncdimmat.shape[0] + 1)

# Prepare plot
fig, axes = plt.subplots(2, 2, figsize=(6, 5))

# Use diverging colormap centered at 0
cmap = plt.get_cmap('seismic')

# Determine max absolute value across all diffs for symmetric color scaling
all_diffs = [diff_A3, diff_A8, diff_T3, diff_T8]
max_abs = max(np.abs(m).max() for m in all_diffs)
norm = Normalize(vmin=-max_abs, vmax=+max_abs)

# Matrix and title setup
diff_matrices = [[diff_A3, diff_A8], [diff_T3, diff_T8]]
titles = [['Wild Type - Act3 Dimer', 'Wild Type - Act8 Dimer'], ['DNA-Bound Wild Type - Act3', 'DNA-Bound Wild Type - Act8']]
residue_indices = [top_residue_indices, bottom_residue_indices]

# Plot each difference map

for i in range(2):  # Rows: full-length, truncated
    for j in range(2):  # Columns: A3, A8
        ax = axes[i, j]
        matrix = diff_matrices[i][j]
        
        # Define limits per row
        if i == 0:
            min_res, max_res = 1, 132
        else:
            min_res, max_res = 1, 196

        if i == 0:  # top row (WT dimers)
            extent = [1, 132, 1, 132]
        else:       # bottom row (DNA dimers)
            extent = [1, 196, 1, 196]

        c = ax.imshow(matrix, cmap=cmap, norm=norm, origin='lower',
                      extent=[min_res, max_res, min_res, max_res])
        ax.set_aspect('auto')  # Fix unequal widths
        ax.set_xlim(min_res, max_res)
        ax.set_ylim(min_res, max_res)
        
        ax.set_title(titles[i][j], fontsize=10)
        ax.tick_params(axis='both', labelsize=9)
        ax.set_xticks(np.linspace(min_res, max_res, 4, dtype=int))
        ax.set_yticks(np.linspace(min_res, max_res, 4, dtype=int))
        ax.tick_params(labelbottom=True)
        
        if j == 0:
            ax.set_ylabel("Residue", fontsize=10)
        if i == 1:
            ax.set_xlabel("Residue", fontsize=10)


# Add colorbar
plt.subplots_adjust(right=0.85, hspace=0.3)
cbar_ax = fig.add_axes([0.88, 0.15, 0.02, 0.7])
cbar = fig.colorbar(c, cax=cbar_ax)
cbar.ax.set_ylabel(' Contact Probability (Wild Type - Mutant)', fontsize=10)
cbar.ax.tick_params(labelsize=9)
ax.set_xlim(extent[0], extent[1])
ax.set_ylim(extent[2], extent[3])

plt.suptitle('Difference Contact Maps (Wild Type - Mutant)', fontsize=12)
plt.savefig('dimer_contactmap_differences.pdf')

# Save figure
#plt.savefig('dimer_mut_contact_maps.pdf')
#plt.savefig('dimer_contact_maps.png', dpi=200)
```
